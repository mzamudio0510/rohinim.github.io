```python
# Dependencies
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
import urllib.request
import requests
import time
```

```python
# Outline the path for chromedriver so splinter can run and read through the URL pages
executable_path = {"executable_path": "chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# First, let's go to the Nasa News URL
url = "https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films"
browser.visit(url)
```

```python
# Scrape Time
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Check out the soup 🥣
print(soup.prettify())
```

```python
# K, now put into tables with pandas 🐼
tables = pd.read_html(url)
tables
```

```python
# Check out the type
type(tables)
```

```python
#Slice and Dice, Make it Nice
disney_pd = tables[4]
disney_pd.columns = ['Type of Film','Title', 'US Release', 'Notes']
disney_pd.head(100)
```

```python
# Count Stuff
disney_pd.count()
```

```python
# Checkout the Nans
disney_pd.isnull().sum()
```

```python
 # Check out the data types, too!
disney_pd.dtypes
```

```python
 #------------------------------------------------#
# It's Cleaning Time, Coderella ♣♣♣
#------------------------------------------------#

# Make a DataFrame and only include the columns you need to use.

disney_pd = disney_pd[['Type of Film', 
                          'Title', 
                          'US Release',
                          ]]

disney_pd = pd.DataFrame(disney_pd)

# Check it out
disney_pd.isnull().sum()
```

```python
 # K, now exit all the NaNs and preview
disney_pd['Type of Film'].dropna()
disney_pd = disney_pd.dropna()
disney_pd.head(100)
```

```python
#Make it a CSV for SQL!
disney_pd.to_csv("WikiDisney.csv", encoding="utf-8", index=False)
```

```python
# Shut it down
browser.quit()
```
