```python
# Dependencies
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
import urllib.request
import requests
import time
```

```python
 # URL of page to be scraped
url = 'https://disneymovieslist.com/list-of-disney-movies/'
```

```python
 # Retrieve page with the requests module
response = requests.get(url)
```

```python
# Check the resposne - 200 means it worked
response
```

```python
# Scrape Time
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Check out the soup 🥣
print(soup.prettify())
```

```python
# K, now put into tables with pandas 🐼
tables = pd.read_html(url)
tables
```

```python
#Slice and Dice, Make it Nice
disney_pd = tables[0]
disney_pd.columns = ['TITLE', 'YEAR']
disney_pd.head(100)
```

```python
# Count Stuff
disney_pd.count()
```

```python
# Checkout the Nans
disney_pd.isnull().sum()
```

```python
 # Check out the data types, too!
disney_pd.dtypes
```

```python
#Make it a CSV for SQL!
disney_pd.to_csv("disneylist.csv", encoding="utf-8", index=False)
```
